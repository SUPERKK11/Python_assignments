# Unpacking 

```python
Unpacking:- Dividing the collection into individual values that is known as unpacking
Syntax:
def f_name(args):
    (code)
f_name(*collection)
Note: numbers of "args" should be same as length of collection
example:
def unpack(a1,a2,a3,a4):
    return a1
unpacked = unpack(*(1,2,3,4))
unpacked
def unpack(*a1):
    return a1
unpacked = unpack(*(1,2,3,4))
unpacked

```

````python
def unpack(a1,a2,a3,a4):
    return a1
unpacked = unpack(*(1,2,3,4))
unpacked
````

````python
def unpack(*a1):
    return a1
unpacked = unpack(*(1,2,3,4))
unpacked
````

````python
def unpack(a1,b1):
    return a1
unpacked = unpack(*{"r":"riya","k":"kashyap"})
unpacked
````

````python
def unpack(a1,b1):
    return a1
unpacked = unpack(*{"r":"riya","k":"kashyap"}.values())
unpacked
````

````python
def unpack(a1,b1):
    return a1
unpacked = unpack(*{"r":"riya","k":"kashyap"}.items())
unpacked
````
