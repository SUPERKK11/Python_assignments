# default values

```python
Passing default values as an arguments
def f_name(a,b,c,d=0,e=" "):
    (code)
f_name(v1,v2,v3)
or 
f_name(v1,v2,v3,v4,v5)
```

````python
def f_name(a,b,c,d=0,e="A"):
    print(a,b,c,d,e)
f_name(1,2,3)
````

````python
def f_name(a,b,c,d=0,e="A"):
    print(a,b,c,d,e)
f_name(1,2,3,4,"k")
````

````python
#create a functioon to add atleast 3 number and max 5 number
def add(a,b,c,d=0,e=0):
    return a+b+c+d+e
add_1=add(1,2,3)
add_2=add(1,2,3,4)
add_3=add(1,2,3,4,5)
print(add_1, add_2, add_3)
````
