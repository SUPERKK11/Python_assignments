```python
Kadanes Algorithm:-
This algorithm help us to find the maximum sum of any continuous subarray in O(n) time complexity
```

#### 1.Brute force

````python
arr=[-2, 1, -3, 4, -1, 2, 1, -5, 4]
n=len(arr)
max_sum=float('-inf')
start = end = 0
for i in range(n):
    current_sum=0
    for j in range(i,n):
        current_sum=current_sum+arr[j]
        if current_sum > max_sum:
            max_sum=current_sum
            start=i
            end=j
print(arr[start:end+1])
````

#### 2.Basic kadanes

````python
arr=[-2,-1,-3,-4,-1,-2,-1,-5,-4]
n=len(arr)
max_sum=float('-inf')
start=end=temp=0
current_sum=0
for i in range(n):
    current_sum=current_sum+arr[i]
    if current_sum > max_sum:
        max_sum=current_sum
        start=temp
        end=i
    if current_sum < 0:
        current_sum=0
        temp=i+1
print(max_sum)
print(arr[start:end+1])
````

#### 3.Standard Kadane's algorithm (DP version)

````python
arr=[-2,-1,-3,-4,-1,-2,-1,-5,-4]
n=len(arr)
start=end=temp=0
max_sum=arr[0]
current_sum=arr[0]

for i in range(1,n):
    if arr[i] > current_sum+arr[i]:
        current_sum=arr[i]
        temp=i
    else:
        current_sum=current_sum+arr[i]
    
    if current_sum > max_sum:
        max_sum=current_sum
        start=temp
        end=i
print(max_sum)
print(arr[start:end+1])
````
