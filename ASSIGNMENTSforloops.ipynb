```python
# ==========================================
# 1. THE FUNCTION (No State / Amnesia)
# ==========================================
def calculate_damage(health, damage):
    new_health = health - damage
    return new_health

print("--- Function Test ---")

# We have to provide the starting health (100) every single time.
# The function does its math, spits out the answer, and forgets everything.
print(f"Attack 1: Player health is now {calculate_damage(100, 40)}")

# If we attack again, it has no memory of the first attack. 
# It just calculates 100 - 40 again.
print(f"Attack 2: Player health is now {calculate_damage(100, 40)}") 

print("\n")
```

```python
# ==========================================
# 2. THE CLASS & OBJECT (State / Memory)
# ==========================================
class Player:
    def __init__(self):
        # This is the "data waiting for you". 
        # When an object is created, it starts with 100 health.
        self.health = 100 

    def take_damage(self, damage):
        # It modifies its OWN data.
        self.health = self.health - damage
        return self.health

print("--- Object Test ---")

# 1. We instantiate the object. It now exists in memory with 100 health.
player_one = Player() 

# 2. We attack it once. 
print(f"Attack 1: Player health is now {player_one.take_damage(40)}")

# 3. We attack it again. Because player_one is a living object holding state,
# it remembers that its health was reduced to 60.
print(f"Attack 2: Player health is now {player_one.take_damage(40)}")
```
