Intermediate termination : making the looping statement to stop or break in between of the execuation that is iteration that is known as intermediate termination.

there are three keywords used in 
.1 break
.2 continue 
.3 pass

Break : it is a keyword that is use to stop the iteration


```python
gf=int(input("enter the number of gf : "))
mango=int(input("enter the number of mangoes : "))
for i in range(1,gf+1):
    if i>mango:
        break
    print("khalo babu ;)")
```

```python
for i in range(1,5):
    print(i)
    if i == 3:
        break
```

```python
n=int(input("enter the number : "))
count=0
for i in range(1,n+1):
    if n % i == 0:
        count=count+1
    if count > 3:
        print("not prime")
        break
        
```

```python
for i in range(1,6):
    if i == 3 and  i == 5 :
        continue
    print(i)
```
