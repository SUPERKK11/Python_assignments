16. Write Java Program to print factorial of a numbers.

```python
#recurrsion logic (back trackting)
def fact(i,n):
    if i < 1:
        print(n)
        return
    return fact(i-1,n*i)
fact(5,1)
```

```python
#recurrsion without backtracking
def fact(i,n,x):
    if i > x:
        print(n)
        return
    fact(i+1,n*i,x)
fact(1,1,5)
```
