```text
1. Find the product of digits of a number(exclude zeroes).
Eg: 45012 
Output: 40 
```

````python
n=45012
product=1
while n>0:
    digit=n%10
    if digit != 0:
        product=product*digit
    n=n//10
product
````

```text
2. Take user input for start and end range and find all the prime numbers in that range.
Create a separate method to find prime number using most efficient way.
Eg: start: 20, end: 40
Output: 23, 29, 31, 37
```

````python
def prime(n):
    if n==1 or n==2 or n==3:
        return n
    if n%2 == 0:
        return
    for i in range(3,int(n),2):
        if n % i == 0:
            return
    return n
        
n1=20
n2=40
for i in range(n1,n2):
    prime_num=prime(i)
    if prime_num is not None:
        print(prime_num,end=" ")
        
````
