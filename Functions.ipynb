````python
def armstrong(n):    
    def count(n):
        count=0
        while n>0:
            count=count+1
            n=n//10
        return count
    def sum(n):
        sum=0
        power=count(n)
        while n>0:
            digit=n%10
            sum=sum+(digit**power)
            n=n//10
        return sum
    if sum(n) == n:
        print("armstrong")
    else:
        print("not armstrong ")
armstrong(153)
    
````

136.Wap to check whether the number is palindrome or not.

````python
def check(n):
    def reverse(n):
        rev=0
        while n>0:
            digit=n%10
            rev=rev*10+digit
            n=n//10
        return rev
    if n == reverse(n):
        print("palindrome number")
    else:
        print("not a palindrome number")
check(10)
        
````

142.wap to check whether the string is anagram or not.

````python
def anagram(str1,str2): 
    if len(str1) != len(str2):
        return False  
    s=set()
    for i in str1:
        if i not in s:
            s.add(i)
        else:
            return False
    for i in str2:
        if i not in s:
            return False
    return True
anagram("saww","swaw")           
        
````
