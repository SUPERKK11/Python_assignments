# Inbuilt function:-
* The function that is already defined by the developer is known as inbuilt function we can use it but we cannot define it .
* Types Inbuilt function:-

1. Utility function - the function which is common for each and every datatype is known as utility  function eg-(len(),dir(),type(),bool(),id()). 

2. On string - the function that we can only apply on string datatype is known as inbuilt function on string.
* lower():- it is used to convert upper case to lower 
* syntax - var_n.lower()
* Upper():- it is used to convert into lower case to upper
* syntax :- var_n.upper()
* islower():- it is used to check lower case value
* syntax :- var_n.islower()
* isupper():- it is used to check upper case value
* syntax:-var_n.isupper()
* swapcase():- it is used to swap the cases
* syntax:- var_n.swapcase()
* title():- first character of each word into upper case
* syntax :- var_n.title()
* isdigit():- used to check the ch is digit or not
* syntax :- var_n.isdigit()
