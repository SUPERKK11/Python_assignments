124.Wap to check whether the number is arm strong or not. 

````python
def armstrong(n):    
    def count(n):
        count=0
        while n>0:
            count=count+1
            n=n//10
        return count
    def sum(n):
        sum=0
        power=count(n)
        while n>0:
            digit=n%10
            sum=sum+(digit**power)
            n=n//10
        return sum
    if sum(n) == n:
        print("armstrong")
    else:
        print("not armstrong ")
armstrong(153)
    
````
