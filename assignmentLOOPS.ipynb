```python
#for prime factors O(n**0.5)
"""
ohk so the logic is we would first check for the even prime numbers that is 2 which also that is  only even prime number that exists 
next step is if the number is completely didvided by the 2 we would either print it or we can store that as per the questions requirement says 
imp : updation in this n=n//2
then move to next odd number prime number logic 
"""
n=int(input("enter the number for prime factor u want to find : "))
factors=[]
while n%2 == 0:
    factors.append(2)
    n=n//2
i=3
while i*i<=n:# this is because prime factors are in pairs and after some time they repeats 
    while n%i == 0 :
        factors.append(i)
        n=n//i
    i=i+2 #odd number updation

if n>1:
    factors.append(n)

print(factors)
```
