```python
L=[1,2,3,4,5]
L1=[]
for i in range(len(L)-1,-1,-1):
    L1.append(L[i])
L1
    
```

4.remove the duplicate char from the string

```python
s=set()
str1="asssssdd"
str2=""
for i in str1:
    s.add(i)
for i in s:
    str2=str2+i
print(str2)
    
```

5.check whether the middle value of tuple is palindrome


```python
n=eval(input("enter the tuple :  "))
if len(n)%2!=0:
    a=n[len(n)//2]
    suma=0
    while a>0:
        digit=a%10
        suma=suma*10+digit
        a=a//10
    if suma == n[len(n)//2] :
        print("palindrome")
    else:
        print("not a palindrome")        
```
