114. Wap to get the following output. 
              S=’kabab is love’ 
              Out={‘kb’:(‘kbb’,3,’bbk’),’is’:(‘s’,1,’s’),’le’:(‘lv’,2,’vl’)} 
                            { 1st+last char:  (consonant,no of consonant,rev of consonant)}

```python
S="kabab is love"
out={}
counter=0
L=()
str1=""
str2=""
for i in S:
    if i == " ":
        L=L+(str2,)
        L=L+(counter,)
        L=L+(str2[-1::-1],)
        out[str1[::len(str1)-1]]=L
        L=()
        str1=""
        str2=""
        counter=0
    else:
        str1=str1+i
        if i not in "aeiouAEIOU":
            str2=str2+i
            counter=counter+1
L=L+(str2,)
L=L+(counter,)
L=L+(str2[::-1],)
out[str1[::len(str1)-1]]=L
out
```

115.Wap to get the following output. 
                In=[100,200,35,40,60] 
                Out=[335,235,400,395,375] (total sum-value)

```python
IN=[100,200,35,40,60]
OUT=[]
for i in IN:
    sum=0
    for j in IN:
        if i != j:
            sum=sum+j
    OUT.append(sum)
OUT
```
